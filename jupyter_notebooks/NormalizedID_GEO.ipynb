```python
df_normalizedID = pd.read_csv('../data_tables/GEOSeries_count_withoutdataset.csv', error_bad_lines=False, low_memory=False,quoting=3)
df_normalizedID
```

```python
df_normalizedID = df_normalizedID.drop(df_normalizedID.columns[[0]], axis=1)
df_normalizedID
```
