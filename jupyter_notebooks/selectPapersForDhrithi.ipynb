```python
# load packages
import numpy as np
import pandas as pd
```

```python
# import data
p = pd.read_csv('../data_tables/metadataMatrix_raw.csv')
p = p[["journal", "pmc_ID", "pmc_date"]].drop_duplicates()
p
```

```python
# select journals
j = ["Nat_Biotechnol", "Nat_Methods", "Nat_Genet", "Genome_Med", "Genome_Biol", 
     "BMC_Syst_Biol", "BMC_Bioinformatics", "BMC_Genomics", "Nucleic_Acids_Res", 
    "PLoS_Comput_Biol", "Bioinformatics"]
# subset data
pfilt = p[p['journal'].isin(j)]
pfilt
```

```python
# get papers only
paps = pfilt['pmc_ID'].drop_duplicates()
paps
```

```python
# write file out
pfilt.to_csv("../data_tables/selectedJournals_Dhrithi.csv", index = False)
```
