```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data_tables/JournalHomeGrid.csv')
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df.columns = ['Rank', 'FullName','journalName','ImpactFactor'] #change header
df
```

```python
df.drop(df.tail(2).index,inplace=True) #remove last two rows
df
```

```python
#round IF
df['ImpactFactor'] = df['ImpactFactor'].astype(float).round(0)
df
```

```python
df['journalName'] = df['journalName'].replace(' ', '_', regex=True).astype(str)
df
```

```python
df_journal = pd.read_csv('../data_tables/OGstats_splitSRAGEO.csv')
df_journal
```

```python
df_journal['journalName'] = df_journal['journalName'].str.upper()
df_journal
```

```python
df_full=pd.merge(df, df_journal, on="journalName")
df_full
```

```python
colsSRA = ['journalName', 'ImpactFactor','origSRA', 'reuseSRA']
dfSRA = pd.DataFrame(df_full, columns = colsSRA)
colsGEO = ['journalName', 'ImpactFactor','origGEO', 'reuseGEO']
dfGEO = pd.DataFrame(df_full, columns = colsGEO)

totals = dfSRA['origSRA']+dfSRA['reuseSRA']
dfSRA.insert(2, "Total", totals, True)
totals = dfGEO['origGEO']+dfGEO['reuseGEO']
dfGEO.insert(2, "Total", totals, True)

dfSRA
```

```python
dfSRA = dfSRA.loc[dfSRA['Total']>=50]
dfSRA
```

```python
dfGEO
```

```python
dfGEO = dfGEO.loc[dfGEO['Total']>=50]
dfGEO
```

```python
dfSRA['PercentageSRA']=(dfSRA['reuseSRA']/dfSRA['Total'])*100
dfSRA
```

```python
dfSRA_sorted=dfSRA.sort_values('PercentageSRA',ascending=False)
dfSRA_sorted
```

```python
print(dfSRA_sorted.to_string())
```

```python
dfGEO['PercentageGEO']=(dfGEO['reuseGEO']/dfGEO['Total'])*100
dfGEO
```

```python
dfGEO_sorted=dfGEO.sort_values('PercentageGEO',ascending=False)
dfGEO_sorted
```

```python
print(dfGEO_sorted.to_string())
```

```python
plt.style.use(['ggplot', 'seaborn-white'])
plt.figure(figsize=(60,20))

ax = sns.catplot(x='ImpactFactor', y='PercentageSRA', data=dfSRA,
            height=4,aspect=5)
sns.despine()

# set axis labels
ax.axes[0,0].set_title("SRA")
ax.axes[0,0].set_xlabel('ImpactFactor', fontsize = 15)
ax.axes[0,0].set_ylabel('Secondary analysis percentage', fontsize = 15)

ax.savefig('IF_SRA.png', dpi = 200, bbox_inches = 'tight', transparent = False)
```

```python
plt.style.use(['ggplot', 'seaborn-white'])
plt.figure(figsize=(60,20))

ax = sns.catplot(x='ImpactFactor', y='PercentageGEO', data=dfGEO,
            height=4,aspect=5)
sns.despine()

# set axis labels
ax.axes[0,0].set_title("GEO")
ax.axes[0,0].set_xlabel('ImpactFactor', fontsize = 15)
ax.axes[0,0].set_ylabel('Secondary analysis percentage', fontsize = 15)
plt.ylim(0,100)

ax.savefig('IF_GEO.png', dpi = 200, bbox_inches = 'tight', transparent = False)
```

```python
def f_impactfactor(df):
    if df['ImpactFactor']>=0 and df['ImpactFactor']<=5:
        val='IF_0_5'
    elif df['ImpactFactor']>=6 and df['ImpactFactor']<=10: 
        val='IF_6_10'
    elif df['ImpactFactor']>=11 and df['ImpactFactor']<=20: 
        val='IF_11_20'
    else:
        val='IF_above20'
    return val
```

```python
dfSRA_sorted['type']=dfSRA_sorted.apply(f_impactfactor,axis=1)
dfSRA_sorted
```

```python
plt.style.use(['ggplot', 'seaborn-white'])
plt.figure(figsize=(60,20))

ax = sns.catplot(x='type', y='PercentageSRA', data=dfSRA_sorted,
            height=4,aspect=5)
sns.despine()

# set axis labels
ax.axes[0,0].set_title("SRA")
ax.axes[0,0].set_xlabel('ImpactFactor', fontsize = 15)
ax.axes[0,0].set_ylabel('Secondary analysis percentage', fontsize = 15)

ax.savefig('IF_SRA.png', dpi = 200, bbox_inches = 'tight', transparent = False)
```

```python
dfGEO_sorted['type']=dfGEO_sorted.apply(f_impactfactor,axis=1)
dfGEO_sorted
```

```python
plt.style.use(['ggplot', 'seaborn-white'])
plt.figure(figsize=(60,20))

ax = sns.catplot(x='type', y='PercentageGEO', data=dfGEO_sorted,
            height=4,aspect=5)
sns.despine()

# set axis labels
ax.axes[0,0].set_title("GEO")
ax.axes[0,0].set_xlabel('ImpactFactor', fontsize = 15)
ax.axes[0,0].set_ylabel('Secondary analysis percentage', fontsize = 15)

ax.savefig('IF_SRA.png', dpi = 200, bbox_inches = 'tight', transparent = False)
```
