```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
import seaborn as sns
```

```python
df_sra = pd.read_csv('../data_tables/sra_technology_species.csv', error_bad_lines=False, low_memory=False,quoting=3)

df_sra
```

```python
df_sra.columns = ['Technology','SRAStudy','Species']
df_sra
```

```python
df_sra = df_sra.drop_duplicates(subset=['SRAStudy'],keep='first')
df_sra
```

```python
df_sra_count = pd.read_csv('../data_tables/SRAStudy_count.csv', error_bad_lines=False, low_memory=False,quoting=3)

df_sra_count 
```

```python
df_sra_count = df_sra_count.drop(df_sra_count.columns[[0]], axis=1)
df_sra_count
```

```python
#df_sra_count = df_sra_count[df_sra_count['Count'] > 1]#reuse counts
#df_sra_count
```

```python
df_combined=pd.merge(df_sra, df_sra_count, on="SRAStudy")
df_combined
```

```python
df_combined = df_combined.drop(df_combined.columns[[0,1]], axis=1)
df_combined
```

```python
df_sra_species=df_combined.groupby('Species').agg({'Count':'sum'}).reset_index().rename(columns={'Species':"Count"})
df_sra_species
```

```python
df_sra_species.columns = ['Species','Count']
df_sra_species
```

```python
Total = df_sra_species['Count'].sum()
print (Total)
```

```python
df_sra_species['Percentage']=(df_sra_species['Count']/63297)*100
df_sra_species
```

```python
df_sra_species = df_sra_species[df_sra_species['Percentage'] > 5]#remove percentage that less than 5
df_sra_species
```

```python
# set aesthetics
plt.style.use(['ggplot', 'seaborn-white'])
plt.figure(figsize=(40,25))

# generate axes
ax = sns.catplot(x='Species', y='Count', data=df_sra_species,
            height=4,aspect=5,kind='bar')
sns.despine()
#plt.ylim(0, 50)

# set axis labels
ax.axes[0,0].set_xlabel('Species', fontsize = 15)
ax.axes[0,0].set_ylabel('Count', fontsize = 15)
plt.xticks(rotation=90)
plt.yscale('log')

#save to .png
ax.savefig('reuse_by_species_SRA.png', dpi = 200, bbox_inches = 'tight', transparent = False)
```

```python
df_combined=df_combined.loc[df_combined['Species'].isin(['Homo sapiens','soil metagenome'])]
df_combined
```

```python
df_human = df_combined[df_combined.Species == 'Homo sapiens']
df_human
```

```python
df_human_reuse = df_human[df_human.Count != 1]
df_human_reuse
```

```python
df_human_reuse['Count']=df_human_reuse['Count']-1
df_human_reuse
```

```python
df_soil = df_combined[df_combined.Species == 'soil metagenome']
df_soil
```

```python
df_soil_reuse = df_soil[df_soil.Count != 1]
df_soil_reuse
```

```python
df_soil_reuse['Count']=df_soil_reuse['Count']-1
df_soil_reuse
```

```python
df_combined_reuse = df_human_reuse.append(df_soil_reuse, ignore_index=True)
df_combined_reuse
```

```python
# set aesthetics
plt.style.use(['ggplot', 'seaborn-white'])
plt.figure(figsize=(40,25))

# generate axes
ax = sns.catplot(x='Species', y='Count', data=df_combined_reuse,
            height=4,aspect=5,kind='boxen')
sns.despine()

# set axis labels
ax.axes[0,0].set_xlabel('Species', fontsize = 20)
ax.axes[0,0].set_ylabel('Count', fontsize = 20)
#plt.yscale('log')

#save to .png
ax.savefig('Reuse_by_species_SRA.png', dpi = 200, bbox_inches = 'tight', transparent = False)
```
