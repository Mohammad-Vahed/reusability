```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd

import seaborn as sns

df = pd.read_csv('OGstats.csv')
```

```python
ax = sns.barplot(x='journalName', y='values', hue='tags', data=df)
plt.xticks(rotation=90)

graph = ax.get_figure()
graph.savefig('figure.png', dpi=1000, bbox_inches = 'tight')
```
