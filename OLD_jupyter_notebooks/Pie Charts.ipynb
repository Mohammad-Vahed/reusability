```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd

import seaborn as sns
sns.set_palette('pastel')

df = pd.read_csv('pmcBasicStats_withMulti.csv')
df
df_clean = df.iloc[1:8,0:11]
df_clean
```

```python
labels = df_clean['x']
#labels

journals=["Bioinformatics", "BMC_Bioinformatics", "BMC_Genomics", "BMC_Syst_Biol", "Genome_Biol", "Genome_Med", "Nat_Biotechnol", "Nat_Methods", "Nucleic_Acids_Res", "PLoS_Comput_Biol"]
for i in journals:
    sizes = df_clean[i]
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels)
    ax1.axis('equal')
    plt.title(i, pad=20)
    plt.tight_layout()
    #plt.show()
    filename = "piechart_" + i + ".png"
    fig = ax1.get_figure()
    fig.savefig(filename, dpi = 1000)
    
```
