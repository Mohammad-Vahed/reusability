```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
import seaborn as sns
```

```python
df_geo_reuse = pd.read_csv('../data_tables/geoReuseCounts.csv', error_bad_lines=False, low_memory=False,quoting=3)
df_geo_reuse
```

```python
df= df_geo_reuse[(df_geo_reuse['ID'].str.startswith('GDS', na=False))| (df_geo_reuse['ID'].str.startswith('GSE', na=False))| (df_geo_reuse['ID'].str.startswith('GPL', na=False))]
df
```

```python
df['reuseCount']=df['reuseCount']+1
df
```

```python
df_geo_sample = pd.read_csv('../data_tables/geo_sample.csv', error_bad_lines=False, low_memory=False,quoting=3)
df_geo_sample
```

```python
df_geo_sample.columns = ['GEOSample', 'GEOPlatform','GEOSeries']
df_geo_sample
```

```python
df_geo_series = pd.read_csv('../data_tables/geo_series.csv', error_bad_lines=False, low_memory=False,quoting=3)
df_geo_series
```

```python
df_geo_series.columns = ['GEOSeries', 'Sample Count','GEODatasets','PubMed ID']
df_geo_series
```

```python
#GEO series
```

```python
df.columns = ['GEOSeries', 'reuseCount']
df
```

```python
df_geoID_series = df_geo_series[['GEOSeries']]
df_geoID_series
```

```python
df_GEOSeries=pd.merge(df, df_geoID_series, on="GEOSeries")
df_GEOSeries
```

```python
df1=pd.concat([df, df_GEOSeries]).drop_duplicates(keep=False)#remove GEO series IDs
df1
```

```python
#GEO Platform ID
```

```python
df1.columns = ['GEOPlatform', 'reuseCount']
df1
```

```python
df.columns = ['GEOPlatform', 'reuseCount']
df
```

```python
df_geoID_platform = df_geo_sample[['GEOPlatform','GEOSeries']].copy()
df_geoID_platform = df_geoID_platform.drop_duplicates(subset='GEOPlatform')

df_geoID_platform
```

```python
df_GEOPlatform=pd.merge(df, df_geoID_platform, on="GEOPlatform")
df_GEOPlatform
```

```python
df_platform_to_series=df_GEOPlatform.drop(columns=['GEOPlatform'])
df_platform_to_series
```

```python
df_platform1=df_GEOPlatform.drop(columns=['GEOSeries'])
df_platform1
```

```python
df2=pd.concat([df1, df_platform1]).drop_duplicates(keep=False)#remove GEO platform IDs
df2
```

```python
#calculate the use by normalizing to GEO series IDs
```

```python
df_new=pd.concat([df_GEOSeries,df_platform_to_series],sort=True)#combine two tables together
df_new
```

```python
df_count=df_new.groupby('GEOSeries').agg({'reuseCount':'sum'}).reset_index().rename(columns={'GEOSeries':"reuseCount"})
df_count
```

```python
df_count.columns = ['GEOSeries', 'Count']
df_count
```

```python
df_count.to_csv('GEOSeries_count_withoutdataset.csv')
```
