```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd

import seaborn as sns
sns.set_palette('pastel')

df = pd.read_csv('pmcBasicStats.csv')
totals = df.sum(axis=1)
df.insert(1, "Total", totals, True)

# remove all-zero rows
clean_df = df[df.Total != 0]

sum_data = clean_df.sum()[2:9]
headers = ['Repository', 'Count']
sum_data
```

```python
# pie chart of workers
labels = df['x']
#labels

journals=["Bioinformatics", "BMC_Bioinformatics", "BMC_Genomics", "BMC_Syst_Biol", "Genome_Biol", "Genome_Med", "Nat_Biotechnol", "Nat_Methods", "Nucleic_Acids_Res", "PLoS_Comput_Biol"]
labels = df['repoName']
for i in journals:
    sizes = df[i]
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels)
    ax1.axis('equal')
    plt.title(i, pad=20)
    plt.tight_layout()
    #plt.show()
    filename = "piechart_" + i + ".png"
    fig = ax1.get_figure()
    fig.savefig(filename, dpi = 1000)
    
```
